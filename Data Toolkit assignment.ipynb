1. What is NumPy, and why is it widely used in Python?
- Numpy is a powerful open-source python  library used for numerical and scientific computing. It provides support for large, multi-dimensional arrays and matrices, along with a collection of high-level mathematical functions to operate on these arrays efficiently.

 **Key Features of NumPy:**

1. N-dimensional Array Object (ndarray)

The core of NumPy is its fast and memory-efficient array object.

It’s more efficient than regular Python lists.

2. Broadcasting

Allows arithmetic operations between arrays of different shapes (without writing loops).

3. Mathematical Functions

Comes with built-in support for linear algebra, Fourier transforms, statistics, etc.

4. Performance

Written in C under the hood, so operations are much faster than native Python.

5. Integration

Works well with other libraries like Pandas, Matplotlib, SciPy, Scikit-learn, TensorFlow, etc.


```python
#example
import numpy as np

# Create an array
a = np.array([1, 2, 3])

# Vectorized operations
b = a * 2  # array([2, 4, 6])

# Matrix multiplication
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
result = np.dot(A, B)
print(result)
```

2. How does broadcasting work in NumPy?
- Broadcasting in NumPy is a powerful concept that allows arrays of different shapes to be used together in arithmetic operations. Rather than explicitly reshaping arrays, NumPy automatically "broadcasts" the smaller array across the larger one so they can have compatible shapes for element-wise operations.

```python
#example
#simple scalar broadcasting
import numpy as np

a = np.array([1, 2, 3])
b = 2

print(a + b)
```

```python
#1D and 2D array broadcasting
a = np.array([[1, 2, 3],
              [4, 5, 6]])

b = np.array([10, 20, 30])

print(a + b)
```

```python
#Incompatible shapes (will raise an error)
a = np.array([[1, 2, 3],
              [4, 5, 6]])

b = np.array([10, 20])

print(a+b)


```

4.  Explain the use of the groupby() method in Pandas.
- The groupby() method in Pandas is used to split a DataFrame into groups based on one or more keys (like column values), apply some operation, and then combine the results.

This is often referred to as the “Split-Apply-Combine” strategy.



```python
#example
#group by one column
import pandas as pd

data = {
    'Department': ['Sales', 'Sales', 'HR', 'HR', 'IT'],
    'Employee': ['A', 'B', 'C', 'D', 'E'],
    'Salary': [50000, 60000, 45000, 48000, 70000]
}

df = pd.DataFrame(data)

# Group by Department and get average salary
grouped = df.groupby('Department')['Salary'].mean()
print(grouped)
```

```python
#group by multiple columns
data = {
    'Team': ['A', 'A', 'B', 'B'],
    'Year': [2020, 2021, 2020, 2021],
    'Points': [25, 30, 45, 50]
}
df = pd.DataFrame(data)

df.groupby(['Team', 'Year'])['Points'].sum()
```

13.  Why is handling missing data important in Pandas?
- Handling missing data in Pandas is super important because missing values can mess up your analysis, visualizations, or machine learning models if not handled properly. Think of it as cleaning your lens before trying to take a clear picture — if the data’s messy, the results will be too.

1. Inaccurate Analysis

Missing values can skew your stats like mean, median, standard deviation, etc.

You might draw the wrong conclusions if you ignore them.

2. Modeling Issues

Most ML algorithms can’t handle NaNs (Not a Number) — they’ll either error out or produce unreliable results.

Filling or dropping missing values is often a required preprocessing step.

3. Plotting Problems

Visualization libraries like Matplotlib, Seaborn, or Plotly may misbehave or skip data when NaNs are present.

4. Data Integrity

Handling missing data ensures your dataset is complete, clean, and consistent, which is vital for any kind of reporting or business decision-making.

17.  Explain the difference between apply() and map() in Pandas.

Best when you want to transform values in a single column/Series.

Often used for value replacement or simple transformations.

 **apply() — For Applying Functions Across Rows/Columns or Elements**

Works on Series and DataFrames.

More flexible and powerful than map().

21.  Why is NumPy’s array slicing faster than Python’s list slicing?
- The reason NumPy’s array slicing is faster than Python’s list slicing comes down to how data is stored and accessed in memory and how NumPy is implemented under the hood. Here's a clear breakdown:

1. Contiguous Memory Allocation

NumPy arrays are stored in a contiguous block of memory.

This allows efficient computation and slicing using pointer arithmetic, with no need to copy data during slicing (in most cases).

Python lists, on the other hand, are arrays of pointers to objects, so slicing involves copying those references one by one.

2. C-backed Implementation

NumPy is written in C, so operations like slicing are done at the compiled level, not in interpreted Python.

This gives a massive speed boost due to low-level optimizations and no Python-level overhead.

3. Views Instead of Copies

NumPy slicing usually returns a view of the original array, not a copy

4. Homogeneous Data

NumPy arrays are homogeneous (all elements are of the same type), which means NumPy can use vectorized operations and memory-efficient layouts.

Python lists are heterogeneous, so each element is a generic Python object, making operations slower.



```python
#quick benchmark
import numpy as np
import time

a = np.arange(1_000_000)
start = time.time()
b = a[100:999000]
print("NumPy slice:", time.time() - start)

lst = list(range(1_000_000))
start = time.time()
b = lst[100:999000]
print("List slice:", time.time() - start)
```

```python
#13. Create a 3D scatter plot using Plotly.
import plotly.express as px
import pandas as pd
import numpy as np

np.random.seed(42)  # For reproducibility
df = pd.DataFrame({
    'X': np.random.rand(100),
    'Y': np.random.rand(100),
    'Z': np.random.rand(100),
    'Color': np.random.rand(100)  # For coloring points
})

fig = px.scatter_3d(df, x='X', y='Y', z='Z', color='Color',
                    title='3D Scatter Plot', labels={'X': 'X-axis', 'Y': 'Y-axis', 'Z': 'Z-axis'})

fig.show()
```
